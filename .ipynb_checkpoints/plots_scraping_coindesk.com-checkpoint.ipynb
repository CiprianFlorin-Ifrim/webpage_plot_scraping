```python
!pip install selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.remote.webelement import WebElement
import time
from PIL import Image

#-------------------------------------------------------------------------------------JUPYTER NOTEBOOK SETTINGS-------------------------------------------------------------------------------------
from IPython.core.display import display, HTML                                    
display(HTML("<style>.container { width:100% !important; }</style>"))     
```

```python
def get_values_of_element(locator_type, locator):
    WebDriverWait(driver, 2).until(EC.presence_of_element_located((locator_type, locator)))
    elements = driver.find_elements(locator_type, locator)
    
    return elements
```

```python
webpage = "https://coindesk.com/data/"

driver = webdriver.Chrome()
driver.implicitly_wait(2) # seconds
driver.get(webpage) 

# scroll down the page to load more currencies/names - 14 scrolls are enough to get all the values
for i in range(0, 2):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)

# find all elements with the class name requested
currencies_names = list()
currency_name_element = "span.typography__StyledTypography-owin6q-0.gtxndB"
currencies_available = get_values_of_element(By.CSS_SELECTOR, currency_name_element) 

# check if any elements were found
if not currencies_available:
    print("No elements found with class name " + currency_name_element)
else:
    for element in currencies_available:
        currencies_names.append(element.text.lower().replace(" ", "-"))
              
print(currencies_names)

# start navigating the page and taking screenshots
for k in range(len(currencies_names)):
    driver.get("https://www.coindesk.com/price/" + currencies_names[k])
    timeframe_xpath_expression = (
                    '//div[@class="Box-sc-1hpkeeg-0 fFbIq date-btn-holder"]/button['
                    '@class="Button__ButtonBase-sc-1sh00b8-0 Button__TextButton-sc-1sh00b8-1 '
                    'hYnMok kjNGnx price-chartstyles__DateButton-sc-1lyescv-2 hHBgMJ"]')
    
    try:
        timeframe_buttons = get_values_of_element(By.XPATH, timeframe_xpath_expression)
        #print(timeframe_buttons)
    except TimeoutException:
        continue
                                                           
    loop_counter = 0
    #print("Size:", len(timeframe_buttons))
    while loop_counter < len(timeframe_buttons):
        try:
            timeframe_buttons[loop_counter].click()
        except StaleElementReferenceException:
            timeframe_buttons = get_values_of_element(By.XPATH, timeframe_xpath_expression)
            timeframe_buttons[loop_counter].click()
            #time.sleep(2)                                                           
                                                                                                                                                                            
        # take screenshot of the page
        filepath = "sample_plots/" + webpage.split("/")[2] + "/"
        screenshot_name = timeframe_buttons[loop_counter].text + "_" + currencies_names[k] + ".png"
        driver.save_screenshot(filepath + screenshot_name)

        # get the dimensions of the graph element
        graph_area_element = "div.price-page-currency-detail"
        graph = WebDriverWait(driver, 2).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, graph_area_element))
        )
        graph_location = graph.location
        graph_size = graph.size

        # open the screenshot and crop it to the dimensions of the element
        image = Image.open(filepath + screenshot_name)

        graph_left   =  graph_location['x']
        graph_top    =  graph_location['y']
        graph_right  =  graph_location['x'] + graph.size['width']
        graph_bottom =  graph_location['y'] + graph.size['height'] - 517

        image = image.crop((graph_left, graph_top, graph_right, graph_bottom))
        image.save(filepath + screenshot_name)

        print("Run " + str(loop_counter) + ":", timeframe_buttons[loop_counter].text, currencies_names[k])
        loop_counter += 1

driver.quit()
```
